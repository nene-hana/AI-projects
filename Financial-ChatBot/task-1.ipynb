Your task is to manually extract key financial data for the last three fiscal years from the 10-K filings of Microsoft, Tesla, and Apple. Following the data collection, you will use Python to analyze this data, focusing on trends and insights that could inform the development of an AI-powered financial chatbot.

```python
import pandas as pd
```

```python
df=pd.read_csv("Financials_Microsoft_Tesla_Apple.csv")
```

Clean & inspect the data

```python
df.info()
```

```python
df.columns 
```

```python
df
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue (Billion $)'].pct_change() * 100
```

```python
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income (Billion $)'].pct_change() * 100
```

```python
df
```

```python
df.fillna(0, inplace=True)
```
