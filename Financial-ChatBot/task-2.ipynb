Basic Chatbot Development

```python
def simple_chatbot(user_query):
    responses = {
        "What is the total revenue?": "Microsoft: $211.91B, Tesla: $81.46B, Apple: $394.33B",
        "How has net income changed over the last year?": "Microsoft: 0%, Tesla: 0%, Apple: 0%",
        "What is the cash flow from operations?": "Microsoft: $89.14B, Tesla: $13.54B, Apple: $111.45B"
    }
    return responses.get(user_query, "Sorry, I can only provide information on predefined queries.")

# Example interaction
query = input("Ask a financial question: ")
print(simple_chatbot(query))
```
